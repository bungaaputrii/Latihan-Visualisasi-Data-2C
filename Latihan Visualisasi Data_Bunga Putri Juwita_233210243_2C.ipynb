# LATIHAN VISUALISASI DATA

```python
import matplotlib.pyplot as plt

#Data Contoh
hari = ['senin','selasa','rabu','kamis','jumat','sabtu']
jumlah_penonton = [100, 175, 207, 190, 120, 250]

#grafiknya
plt.figure(figsize=(9,5))
plt.bar(hari,jumlah_penonton, color='y')
plt.title('Bar Chart Jumlah Penonton Per Hari')
plt.xlabel('hari')
plt.ylabel('Jumlah Penonton')
plt.grid(False)
plt.show()
```

Data diatas merupakan data visualisasi data dari jumlah penonton per hari.Menggunakan library matplotlib dengan codingan yang tertera seperti diatas.

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Data Contoh
usia = [20, 22, 21, 23, 24, 25]
ipk = [3.75, 2.00, 4.00, 3.59, 3.76, 3,97]

#Histogram
plt.figure(figsize=(7,5))
sns.histplot(usia, kde=True, color='blue')
plt.title('Distribusi Usia')
plt.xlabel('Usia')
plt.ylabel('IPK')
plt.grid(True)
plt.show()
```

Merupakan visualisasi data menggunakan library seaborn dari distribusi usia dengan indeks prestasi kumulatif.

```python
import matplotlib.pyplot as plt

# Data contoh
hari = ['H1', 'H2', 'H3', 'HA', 'H5', 'H6', 'H7', 'H8', 'H9', 'H10']
jumlah_penjualan = [100, 128, 150, 240, 170, 245, 165, 175, 235, 275]


#grafiknya
plt.figure(figsize=(5, 4))
plt.pie(jumlah_penjualan, labels=hari, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Penjualan per hari')
plt.axis('equal') # Agar Lingkaran berbentuk Lingkaran sempurna
plt.show()
```

Merupakan visualisasi data menggunakan library matplotlib dari chart penjualan per hari.

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Data Contoh
usia = [17, 18, 19, 20, 22, 24, 36, 15, 13, 14, 12, 30, 29, 21, 23, 25]
tinggi_badan = [150, 165, 147, 166, 167, 170, 180, 167, 158, 154, 155, 154, 163, 167, 181, 161]

#Scatter plot #1
plt.figure(figsize=(6,4))
sns.scatterplot(x=usia, y=tinggi_badan)
plt.xlabel('usia')
plt.ylabel('tinggi badan')
plt.show()

#Scatter plot #2
plt.figure(figsize=(6, 4))
sns.regplot(x=usia, y=tinggi_badan)
plt.xlabel('usia')
plt.ylabel('tinggi badan')
plt.grid(True)
plt.show()
```

Merupakan visualisasi data menggunakan library seaborn dari data tinggi badan dan usia.
